# Train and Test

## Train and Test df's

```python
# Create train set
train = values8103_copy[values8103_copy['ds'] < '2021-08-03']
test = values8103_copy[values8103_copy['ds'] > '2021-08-03']
train
```

```python
# ds to datetime
train['ds'] = pd.to_datetime(train['ds'])
#train.info()

# Remove Weekends
train_dayofweek = train[train['ds'].dt.dayofweek < 5]

train_dayofweek
```

## Model Forecast

```python
# Set the uncertainty interval to 95% (the Prophet default is 80%)
Model = Prophet(interval_width=0.8, daily_seasonality = True)

# Fit the timeseries to Model
Model.fit(train)

# Use make_future_dataframe() with a monthly frequency and periods = 36 for 3 years
future_dates = Model.make_future_dataframe(periods=116, freq='D')
future_dates = future_dates[future_dates['ds'].dt.dayofweek < 5]

# Predict the values for future dates and take the head of forecast
forecast = Model.predict(future_dates)

preds = (forecast['ds'] > '2021-08-03')
preds = forecast.loc[preds]
preds = preds[['ds','yhat']]
preds = pd.DataFrame(preds)
preds['SecuritiesCode'] = 8103
preds['Date'] = preds['ds']
preds['Target'] = preds['yhat']
preds = preds.drop(['yhat','ds'], axis=1)
#pivot = preds.pivot(index=None, columns=['Date', 'SecuritiesCode'], values='Target')
#full_preds = pd.concat([pivot,full_preds], axis=1)

preds
```
